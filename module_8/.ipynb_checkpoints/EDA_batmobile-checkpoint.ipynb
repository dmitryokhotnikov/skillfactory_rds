```python
def categorical_info(column_name, train, rotate=90, x_shift_mult=0, percentage=True):
    """Рисует распределение категориальных признаков и отображает их процентное содержание"""

    ax = sns.countplot(x=column_name,
                       data=train,
                       order=train[column_name].value_counts().index, orient="h")
    if percentage:
        total = len(train[column_name])
        for p in ax.patches:
            percentage = '{:.1f}%'.format(100 * p.get_height()/total)
            x = p.get_x() + x_shift_mult*p.get_width()
            y = p.get_y() + p.get_height()+8
            ax.annotate(percentage, (x, y))
    plt.xticks(rotation=rotate)


def numeric_info(train, bins=10):
    """Печатает описательные статистики числового признака, а также рисует его распределение"""

    print(train.describe()[["min", "max", "25%", "50%", "75%", "std"]])
    sns.distplot(train, bins=bins)
```

## engineDisplacement (объем двигателя)

**Числовой** признак, представляет собой объем двигателя в литрах. Есть пара исключений - электродвигатель (undefined LTR) и гибридный двигатель (0.7 LTR). Таких примеров очень мало (5 в трейне и 1 в тесте), кроме того, значения мощности в л.с. примерно одинаковы и равно 170 л.с. или 126кВт. Это значение мощности может создаваться бензиновыми двигателями различного объема, возьмем в качестве замены двигатель объемом 2 литра. 

```python
def engine_disp_change(data):
    data['engineDisplacement'] = data['engineDisplacement'].str.split().apply(
        lambda s: s[0])
    data.loc[data.engineDisplacement.isin(
        ['undefined', '0.7']), 'engineDisplacement'] = '2.0'
    data['engineDisplacement'] = data['engineDisplacement'].apply(
        lambda s: float(s))
    return data
```

```python
train = engine_disp_change(train)
test = engine_disp_change(test)
```

```python
numeric_info(train.engineDisplacement, bins=15)
```

```python
numeric_info(test.engineDisplacement, bins=15)
```

Как видим, распределения тренировочных и тестовых данных очень похожи

# Подготовка данных для модели CatboostRegressor

Опишем подробно подготовку табличных данных для модели CatboostRegressor

```python
# загрузим тренировочные и тестовые данные
train = pd.read_csv("train.csv")
test = pd.read_csv("test.csv")
```

```python
def engine_disp_change(data):
    data['engineDisplacement'] = data['engineDisplacement'].str.split().apply(
        lambda s: s[0])
    data.loc[data.engineDisplacement.isin(
        ['undefined', '0.7']), 'engineDisplacement'] = '2.0'
    data['engineDisplacement'] = data['engineDisplacement'].apply(
        lambda s: float(s))
    return data
```

```python
# engineDisplacemment
train = engine_disp_change(train)
test = engine_disp_change(test)
```

```python
# enginePower
test.enginePower = test.enginePower.map(lambda x: float(x.split()[0]))
train.enginePower = train.enginePower.map(lambda x: float(x.split()[0]))
```

```python
# modelDate
test['modelDate'] = 2021 - test['modelDate']
train['modelDate'] = 2021 - train['modelDate']
```

```python
# name
train.name = train.name.map(lambda x: (x.split()[0]))
test.name = test.name.map(lambda x: (x.split()[0]))
```

```python
# productionDate
test['productionDate'] = 2021 - test['productionDate']
train['productionDate'] = 2021 - train['productionDate']
```

```python
# Владельцы
train.loc[train['Владельцы'].isna() == True, 'Владельцы'] = '3 или более'
```

```python
# Добавим новые фичи

# пробег в год
train['mileagePerYear'] = train['mileage']/train['productionDate']
test['mileagePerYear'] = test['mileage']/test['productionDate']

# разделим пробег на группы, исходя из того, что в среднем авто (данные Автостат) проходит за год 17000 км
# каждая группа будет грубо (не всегда пробег правдив) отображать количество лет эксплуатации.

test["exp_year"] = test.mileage.apply(
    lambda x: x//17000 if x <= 400000 else 400000//17000+1)
train["exp_year"] = train.mileage.apply(
    lambda x: x//17000 if x <= 400000 else 400000//17000+1)
```

```python
# Логарифмирование признаков
# ["price", "modelDate", "productionDate", "mileage"]#
log_features = ["price"]

for feature in log_features:
    try:
        train[feature] = np.log(train[feature])
        test[feature] = np.log(test[feature])
    except:
        continue
```

```python
# посмотрим на корреляции числовых признаков
numerical_features = ['engineDisplacement', 'enginePower', 'mileage',
                      'modelDate', 'exp_year', 'productionDate', 'mileagePerYear']
sns.set(font_scale=1)
plt.subplots(figsize=(9, 9))
sns.heatmap(train.loc[:, numerical_features].corr(), square=True,
            annot=True, fmt=".3f", linewidths=0.1, cmap="RdBu", vmin=-1.0, vmax=1.0)
```

Видим, что признаки productionDate и modelDate, а также exp_year и mileage имеют большие значения коэффициента корреляции, поэтому сначала было принято решение удалить их. Но проведенные эксперименты показали, что участие этих признаков в модели дает прирост качества модели. 

```python
# удалим неиспользуемые столбцы
unuse_cols = ["description", "sell_id",
              "vehicleConfiguration", "Владение", "Руль"]
test.drop(unuse_cols, axis=1, inplace=True)
train.drop(unuse_cols, axis=1, inplace=True)
```

```python
# сохраним данные
train.to_csv("train_catboost_1.csv", index=False)
test.to_csv("test_catboost_1.csv", index=False)
```
