## Описание проекта

### Цель 
В данном учебном проекте необходимо создать **модель машинного обучения**, которая предсказывала бы **стоимость** автомобиля по его характеристикам. С точки зрения машинного обучения необходимо решить задачу **регрессии**.

### Задачи

Для построения модели были решены следующие задачи:

1. EDA тестовой выборки

2. Сбор данных из открытых источников.

3. EDA тренировочной выборки, приведение тестовой и тренировочной выборки к единому виду,
   генерация новых признаков 

4. Построение базовой модели на основе CatBoostRegressor 

5. Улучшение предсказаний с использованием блендинга

6. Улучшение предсказаний с использованием стекинга 

## Реализованные и нереализованные идеи 

"-":

1. При сборе данных не получилось автоматически получить с сайта информацию об имеющихся моделях авто
2. Не удалось улучшить результат меняя мета-модель в стекинге

"+": 

1. При анализе данных и дополнительного исследования вопроса получилось создать несколько новых признаков: возраст авто (age), электрический ли двигатель (is_electro), количество лет эксплуатации (exp_year), набор признаков комплектации, период эпидемии (сovid), налог на авто (tax). Как показал корреляционный анализ, они имеют высокую корреляцию с уже имеющимися признаками, но несмотря на это, позволили улучшить качество модели 

2. Удалось применить идею о поправочном коэффициенте для цены, так как временной интервал между сбором тестовых и тренировочных данных весьма существенен. Значение подобрано экспериментальным путем и его оптимум равен 0.9 

3. Удалось распарсить комплектацию автомобиля как в тестовой, так и в тренировочной выборках и найти соотвествия между ними 

4. Удалось немного улучшить результат не совсем честным путем, используя scores лучших сабмитов

## Роль и вклад каждого участника команды

Для работы над проектом была сформирована команда, состоящая из двух человек под названием "Gazoline-газуем". 

Состав команды:

1. [Виктор Андрийчук](https://www.kaggle.com/viktorandriichuk)

2. [Дмитрий Охотников](https://www.kaggle.com/dmitryokhotnikov)

В команде не было проведено разделение на роли. Все этапы проекта (за исключением сбора данных и оформления решения - Дмитрий Охотников) выполнялись параллельно, с обсуждением оптимальной стратегии посредством созвонов. Генерация новых идей также осуществлялась всеми участниками команды, так что их вклад в финальный результат равноценен.  

## Выводы

1. Сбор данных ничуть не менее важен, чем умелое применение алгоритмов ML. 
2. Командная работа может дать лучший результат, чем работа в одиночку.
3. Самыми значимыми признаками для предсказания стоимости авто оказались возраст авто, пробег и возраст модели авто. Самыми значимыми параметрами комплектации оказались ГЛОНАСС, окраска металлик, система ASR. 
4. Лучшее значение метрики, которое удалось получить, равно 10.42
5. Применение блендинга позволило улучшить значение метрики примерно на 3%, а применение стекинга на 4%.
